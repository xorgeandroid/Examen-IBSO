Ejercicio 1:
1.	Crea un diccionario llamado datos_climaticos donde cada clave es el nombre de una ciudad y el valor es una lista de temperaturas (en grados Celsius) registradas a lo largo de una semana.
2.	Para cada ciudad, calcula la temperatura promedio, la temperatura máxima y la mínima de la semana.
3.	Determina cuál fue la ciudad con la temperatura promedio más alta y la más baja durante la semana.


```python
import pandas as pd

# Diccionario con datos climáticos
datos_climaticos = {
    "Ciudad_1": [22, 24, 19, 21, 23, 20, 18],
    "Ciudad_2": [30, 32, 31, 29, 28, 27, 26],
    "Ciudad_3": [15, 17, 16, 14, 13, 12, 11]
}

# Convertir el diccionario en un DataFrame
df = pd.DataFrame(datos_climaticos)
```

```python
promedios = df.mean()
maximas = df.max()
minimas = df.min()
```

```python
resultados = pd.DataFrame({
    "Promedio": promedios,
    "Máxima": maximas,
    "Mínima": minimas
})

resultados
```

```python
# @title Promedio

from matplotlib import pyplot as plt
resultados['Promedio'].plot(kind='line', figsize=(8, 4), title='Promedio')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
ciudad_max_promedio = promedios.idxmax()
ciudad_min_promedio = promedios.idxmin()
```

```python
print(f"La ciudad con la temperatura promedio más alta es {ciudad_max_promedio} con {promedios[ciudad_max_promedio]}°C")
print(f"La ciudad con la temperatura promedio más baja es {ciudad_min_promedio} con {promedios[ciudad_min_promedio]}°C")
```
